```python
# Visualization by Gender:
```

```python
#Total amount spend(gender wise):
Total_spend_by_gender=df1.groupby('Gender')['Total Spend'].sum().reset_index()
Total_spend_by_gender
```

```python
#Visualization in bar plot:
plt.figure(figsize=(10,6))
plt.bar(Total_spend_by_gender['Gender'],Total_spend_by_gender['Total Spend'],color=['violet','red'])
plt.title('Total_spend_by_gender')
plt.xlabel('Gender')
plt.ylabel('Total Spend')
plt.show()
```

```python
#Total amount spend on items(gender wise)
items_purchased_by_gender=df.groupby('Gender')['Items Purchased'].sum().reset_index()
items_purchased_by_gender
```

```python
#Visualization in bar plot:
sns.catplot(x='Gender',y='Items Purchased',kind='bar',data=df1,palette='viridis')
plt.title('Number Of Items Purchased By Gender')
plt.xlabel('Gender')
plt.ylabel('Item Purchased')
plt.show()
```

```python
# Calculate average sattisfaction levels:
satisfaction_mapping={'Satisfied':3, 'Neutral':2, 'Unsatisfied':1}

df1['Satisfaction Scores']=df1['Satisfaction Level'].map(satisfaction_mapping)
df2=df1.copy()
df2
```

```python
#Gender wise average satisfaction levels:
avg_satisfaction_by_gender=df2.groupby('Gender')['Satisfaction Scores'].mean().reset_index()
avg_satisfaction_by_gender
```

```python
#Graphical representation of gender wise average satisfaction levels:
sns.catplot(x='Gender',y='Satisfaction Scores',kind='bar',data=avg_satisfaction_by_gender)
plt.title('Average Satisfaction Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Satisfaction Score')
plt.show()
```

```python
# Visualization by Age
```

```python
#Total amount spend(age wise):
Total_Spend_By_Age=df.groupby('Age')['Total Spend'].sum().reset_index()
Total_Spend_By_Age
```

```python
#graphical representation of total amount spend(age wise):
plt.figure(figsize=(10,6))
plt.pie(Total_Spend_By_Age['Total Spend'],labels=Total_Spend_By_Age['Age'],autopct='%1.1f%%')
plt.title('Total Spend By Age')
plt.show()
```

```python
#Total spend on items by age:
items_purchased_by_age=df.groupby('Age')['Items Purchased'].sum().reset_index()
items_purchased_by_age
```

```python
#Graphical representation of Total spend on items by age(line graph):
plt.figure(figsize=(10,6))
plt.plot(items_purchased_by_age['Age'],items_purchased_by_age['Items Purchased'],color='blue')
plt.title('Items Purchased By Age')
plt.xlabel('Age')
plt.ylabel('Items Purchased')
plt.show()
```

```python
#Membership type by age:
membership_by_age=df1.groupby(['Age','Membership Type']).size().unstack().fillna(0)
membership_by_age
```

```python
#Graphical representation of membership type by age:
membership_by_age.plot(kind='bar',stacked=True, figsize=(10,6),color=['yellow','blue','red'])
plt.title('Membership Type Distribution by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Membership Type')
plt.show()
```

```python
# Customers based on membership type:
customer_count_by_membership=df1['Membership Type'].value_counts().reset_index()
customer_count_by_membership.columns=['Membership Type','Customer Count']
customer_count_by_membership
```
